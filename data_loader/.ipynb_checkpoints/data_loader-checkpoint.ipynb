```python
from pymongo import MongoClient
uri = "mongodb://bitcoin_tracing:bt2018@localhost:27017/BT" 
client = MongoClient(uri)
print(client.list_database_names())
```

```python
db = client['BT']
collect = db['papers']
```

```python
collect.insert_one({'foo': 'bar'})
post1 = collect.find_one({'foo': 'bar'})
print(post1)
```

```python
mixer_collection = db['mixer']
tager_collection = db['tager']
```

```python
import csv
with open('Mixer.csv', newline='') as csvfile:

  # 讀取 CSV 檔案內容
  rows = csv.reader(csvfile)

  # 以迴圈輸出每一列
  for row in rows:
    print(row)
    mixer_collection.insert_one({
        'name':row[0],
        'address':row[1],
        'label':row[2],
        'owner':row[3]
    })
```

```python
test = mixer_collection.find_one({'address':'17S5PQpYv7TPkqHLvjigefiZGz2LvjkCro'})
print(test)
```

```python
import csv
from tqdm import tqdm
with open('dataset_IEEEBlockchain2018.csv', newline='', encoding="utf-8") as csvfile:

  # 讀取 CSV 檔案內容
  rows = csv.reader(csvfile)

  # 以迴圈輸出每一列
  for row in tqdm(rows):
    tager_collection.insert_one({
        'name':row[0],
        'address':row[1],
        'class':row[2],
        'owner':row[3]
    })
```
